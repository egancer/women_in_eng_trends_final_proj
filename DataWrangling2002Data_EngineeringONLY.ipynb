```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import descartes
from shapely.geometry import Point, Polygon
%matplotlib inline
```

```python
uni_df = pd.read_csv("ef2002cp.csv")
```

```python
uni_df
```

"EFRACE15" is total Men "EFRACE16" is total Women "LINE" code 8 is total full time undergraduates "CIPCODE" 14 is engineering

```python
uni_df1 = uni_df[['unitid', 'efciplev', 'cipcode', 'efrace15', 'efrace16']]
```

```python
uni_df1.head()
```

```python
uni_df2 = uni_df1.loc[uni_df1["efciplev"].isin([202])]
```

```python
uni_df2
```

```python
uni_df3 = uni_df2[['unitid', 'efrace15', 'efrace16']]
```

```python
uni_df4 = uni_df3.groupby(['unitid']).sum()
```

```python
uni_df4
```

```python
uni_df5 = uni_df4.reset_index()
uni_df5
```

```python
uni_df5['W/M_Ratio'] = uni_df5['efrace16']/uni_df5['efrace15']
uni_df5
```

```python
uni_df5.rename(columns={'unitid':'UNITID'}, inplace=True)
uni_df5.head()
```

# Bring in University Location Data

```python
UniLoc = pd.read_csv("UniversityID_Lat_Long_ONLY.csv")
```

```python
UniLoc
```

JOIN!

```python
df = pd.merge(uni_df5, UniLoc, on = "UNITID")
df
```

Change to GEO Dataframe

```python
crs = {'init':'epsg:4326'}
geometry = [Point(xy) for xy in zip(df["LONGITUDE"], df["LATITUDE"])]
geometry[:3]
geo_df = gpd.GeoDataFrame(df, crs = crs, geometry = geometry)
geo_df.head()
```

```python
geo_df.to_csv('geo_df2002.csv', index = False)
```
