## I have to write a function that will populate a new column within the dataframe
### Q1 is upper left hand quadrent - Lat 50-37.5 deg, Long 125-95 deg
### Q2 is upper right hand quadrent - Lat 50-37.5 deg, Long 95-65
### Q3 is lower left hand quadrent - Lat 37.5-25 deg Long 125-95  
### Q4 is lower right hand quadrent - Lat 37.5-25 Long 95-65


First I will write it as two columns with "North" "South" and "East" "West"

```python
len(df['LATITUDE'])
```

```python
# create a list to store new data
new_var = []

#for each row in the column:
for row in df['LATITUDE']:
    #if value greater than 37.5 and less than or equal to 50
    if row > 37.5:
    #append N for North
        new_var.append("N")
    #otherwise
    else:
        #append S for South
        new_var.append("S")

# create a column from the list
df['NorthSouth'] = new_var
```

```python
df
```

```python
another_var = []

#for each row in the column:
for row in df['LONGITUDE']:
    #if value greater than 37.5 and less than or equal to 50
    if row > -95:
    #append E for East
        another_var.append("E")
    #otherwise
    else:
        #append W for West
        another_var.append("W")

# create a column from the list
df['EastWest'] = another_var
df
```

```python
def get_quad(row):
    if(row['EastWest'] == 'W') and (row['NorthSouth'] == 'N'):
        return 'NW'
    elif (row['EastWest'] == 'W') and (row['NorthSouth'] == 'S'):
        return 'SW'
    elif (row['EastWest'] == 'E') and (row['NorthSouth'] == 'N'):
        return 'NE'
    else:
        return 'SE'

df['Quadrant'] = df.apply(get_quad, axis = 1)
```
